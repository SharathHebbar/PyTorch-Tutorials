```python
import torch
import torch.nn as nn
```

```python
image_size = 384 * 384
patch_size = 16 * 16
num_words = image_size // patch_size
```

```python
# Mentioned in the paper 768 and 16
patch_embed = nn.Conv2d(in_channels=3, out_channels=768, kernel_size=(16, 16), stride=(16, 16))
```

```python
x = torch.randn(1, 3, 384, 384)
x1 = patch_embed(x)
```

```python
# flatten 24 * 24 => 576
# dimension 2
x1 = x1.flatten(2)
x1.shape
```
