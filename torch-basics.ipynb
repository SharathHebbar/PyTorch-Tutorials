```python
import torch
```

```python
# Concate tensors
# 1st dim must match
a = torch.randn(2,5)
b = torch.randn(3,5)
torch.cat([a, b]) # concate in 0th dim
```

```python
# the 0th dim must match
a = torch.randn(2, 3)
b = torch.randn(2, 5)
torch.cat([a, b], 1) # dim 1
```

```python
# Squeezing tensors
# adding dim to tensor

a = torch.tensor([1, 2, 3, 4])
b = torch.unsqueeze(a, 0)
b, b.shape
```

```python
c = torch.unsqueeze(a, 1)
c, c.size
```

# auto grad

```python
x = torch.tensor([1., 2., 3.], requires_grad=True)
y = torch.tensor([4., 5., 6.], requires_grad=True)
z = x+y
z
```

```python
 z.grad_fn
```

```python
s = z.sum()
print(s, s.grad_fn)
```

```python
s.backward()
```

```python
x.grad
```
