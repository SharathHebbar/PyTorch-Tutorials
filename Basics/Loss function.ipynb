```python
import torch
import torch.nn as nn
```

```python
pred = torch.randn(4, 5)
label = torch.randn(4, 5)
```

# MSE

```python
mse = nn.MSELoss(reduction='none')
loss = mse(pred, label)
loss
```

```python
mse = nn.MSELoss(reduction='mean')
loss = mse(pred, label)
loss
```

```python
mse = nn.MSELoss(reduction='sum')
loss = mse(pred, label)
loss
```

```python
((pred - label)**2).mean()
```

# BCE

```python
pred
```

```python
label = torch.zeros(4, 5).random_(0,  2)
label
```

```python
sigmoid = nn.Sigmoid()
```

```python
bce = nn.BCELoss(reduction='none')
bce(sigmoid(pred), label)
```

```python
bce = nn.BCELoss(reduction='mean')
bce(sigmoid(pred), label)
```

```python
bce = nn.BCELoss(reduction='sum')
bce(sigmoid(pred), label)
```

#BCE with Logits

```python
bcel = nn.BCEWithLogitsLoss(reduction='none')
bcel(pred, label)
```

```python
bcel = nn.BCEWithLogitsLoss(reduction='mean')
bcel(pred, label)
```

```python
bcel = nn.BCEWithLogitsLoss(reduction='sum')
bcel(pred, label)
```

```python
import numpy as np
x = pred.numpy()
y = label.numpy()
```

```python
loss_vals = []
for i in range(len(y)):
  batch_loss = []
  for j in range(len(y[i])):
    if y[i][j] == 1:
      loss = -np.log(x[i][j])
    else:
      loss = -np.log(1 - x[i][j])
    batch_loss.append(loss)
  loss_vals.append(batch_loss)

```

```python
#sigmoid
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
x = sigmoid(x)
```
